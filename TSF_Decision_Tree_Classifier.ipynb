**Importing all the libraries**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

**Loading the iris dataset**

```python
iris_data = datasets.load_iris()
train_x, test_x, train_y, test_y = train_test_split(iris_data.data, iris_data.target, random_state = 1)
```

**Building our decision Tree Classifier model**

```python
dt_clf = DecisionTreeClassifier()
dt_clf.fit(train_x, train_y)
```

**Using Graphviz to visualize it graphically**

```python
import graphviz
dot_data = export_graphviz(dt_clf, out_file=None,
                           feature_names = iris_data.feature_names,
                           class_names = iris_data.target_names,
                           filled=True, rounded=True,
                           special_characters = True)
graph = graphviz.Source(dot_data)
graph
```

**Evaluating The Model**

```python
pred_y = dt_clf.predict(test_x)
```

```python
from sklearn.metrics import mean_absolute_error
print("Mean Absolute Error:", mean_absolute_error(test_y,pred_y))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y,pred_y)
```
